```python
import requests
import pandas as pd
import json
```

```python
def get_eia_data_from_api(URL, file_name_with_extension):
    api_url = URL
    response = requests.get(api_url)
    data = response.json()
    df = pd.read_json(json.dumps(data['response']['data']))
    df = df.astype({"value": float})
    df.to_csv(file_name_with_extension, index=False)
    return df
    
```

```python
def get_eia_data_from_api_elec(URL, file_name_with_extension):
    api_url = URL
    response = requests.get(api_url)
    data = response.json()
    df = pd.read_json(json.dumps(data['response']['data']))
    df = df.astype({"price": float})
    df.to_csv(file_name_with_extension, index=False)
    return df
```

```python
elec =get_eia_data_from_api_elec("https://api.eia.gov/v2/electricity/retail-sales/data?api_key=SECRET_KEY&data[0]=price&start=2019-01-01&end=2022-07-31", "retail_electricity.csv") 
```

```python
conn = sqlite3.connect('eia_data.db')
c = conn.cursor()
```

```python
natgas.to_sql('natgas', conn, if_exists='replace', index = False)
pet.to_sql('petroleum', conn, if_exists='replace', index = False)
elec.to_sql("electricity", conn, if_exists='replace', index = False)
```
